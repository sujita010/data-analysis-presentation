# **Box Plot: Age Distribution by Target**

```python
# Crete Box Plot of Age

sns.boxplot(data=df, x="target", y="age")
plt.title('Age Distribution by Target')
plt.xlabel('Target')
plt.ylabel('Age')
plt.show()
```



1. Patients with heart disease (target = 1) tend to be slightly younger on average compared to those without heart disease.

2. There is a wider spread in the age of patients without heart disease, while those with heart disease are more concentrated in the mid-age range.





# **Bar Plot: Heart Disease Frequency by Sex**

```python
pd.crosstab(df.sex,df.target).plot(kind="bar",figsize=(15,6),color=['red','green' ])
plt.title('Heart Disease Frequency for Sex')
plt.xlabel('Sex (0 = Female, 1 = Male)')
plt.xticks(rotation=0)
plt.legend(["Haven't Disease", "Have Disease"])
plt.ylabel('Frequency')
plt.show()
```

1. Males (sex = 1) are more frequently represented in both categories, but especially among those with heart disease.

2. This suggests a higher prevalence or risk of heart disease in males compared to females in this dataset.

# **Scatter Plot: Age vs Max Heart Rate by Disease**

```python
plt.scatter(x=df.age[df.target==1], y=df.thalach[(df.target==1)], c="red")
plt.scatter(x=df.age[df.target==0], y=df.thalach[(df.target==0)])
plt.legend(["Disease", "Not Disease"])
plt.xlabel("Age")
plt.ylabel("Maximum Heart Rate")
plt.show()
```

1. Individuals with heart disease generally achieve lower maximum heart rates compared to those without heart disease.

2. Younger patients tend to have higher maximum heart rates overall, but the distinction between target classes becomes clearer as age increases.

# **Histogram: Age Distribution**


```python
df['age'].hist(bins=20)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

1. Most patients fall between the ages of 40 to 60.

2. This reinforces the understanding that heart disease risk increases with age but is particularly concentrated in middle-aged individuals.

# **Heatmap: Correlation Matrix**

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Between Attributes")
plt.show()
```

1. thalach (maximum heart rate) is negatively correlated with heart disease (target).

2. cp (chest pain type) and sex show moderate positive correlations with heart disease.

3. Features like chol (cholesterol) and fbs (fasting blood sugar) show weaker correlations

# **Count Plot: Heart Disease by Gender**

```python
sns.countplot(x='sex', hue='target', data=df)
plt.title('Heart Disease by Gender')
plt.xlabel('Sex (0 = Female, 1 = Male)')
plt.ylabel('Count')
plt.legend(title='Heart Disease')
plt.show()
```

1. Again confirms that males are more affected by heart disease than females.

2. Among females, the proportion with heart disease appears lower.



# **Scatter Plot: Max Heart Rate vs Age (with Hue = Target)**



```python
sns.scatterplot(x='age', y='thalach', hue='target', data=df)
plt.title('Max Heart Rate vs Age (by Heart Disease)')
plt.xlabel('Age')
plt.ylabel('Max Heart Rate')
plt.show()
```

1. Shows a clear inverse trend: as age increases, maximum heart rate decreases.

2. The color separation (by target) again emphasizes that people with lower max heart rate tend to have heart disease.
